```python
#I3D
from umap import UMAP
import plotly.express as px
import matplotlib.pyplot as plt
from pyod.models.cof import COF
from pyod.utils.example import visualize
from pyod.utils.data import evaluate_print
from pyod.utils.data import generate_data
from torchvision import datasets, transforms
from pyod.utils.utility import precision_n_scores
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import roc_auc_score
import numpy as np

from sklearn.model_selection import train_test_split
norm_data = np.loadtxt('/content/drive/MyDrive/normal_I3D.csv', delimiter=',')
crash_data = np.loadtxt('/content/drive/MyDrive/crash_I3D.csv', delimiter=',')

# norm_data.shape (2999,1024) crash_data.shape (1500,1024)
# crash anomaly is too many so, use only 300 of crash data (10% of normal data)
crash_data = crash_data[0:100, :]

# split 70% for training 30% for testing
N = norm_data.shape[0]
N_tr = np.round(N * .7).astype(int)
train_data = norm_data[:N_tr, :]
test_data = norm_data[N_tr:, :]

train_label = np.zeros(train_data.shape[0])#,dtype="U4")
#train_label[:] = 'normal'
test_label = np.zeros(test_data.shape[0])#,dtype="U4")
#test_label[:] = 'normal'

C = crash_data.shape[0]
C_tr = np.round(C * .7).astype(int)
train_data = np.concatenate([train_data, crash_data[:C_tr, :]])
tmp = np.ones(crash_data[:C_tr, :].shape[0])
#tmp[:] = 'anomaly'
train_label = np.concatenate([train_label, tmp])

test_data = np.concatenate([test_data, crash_data[C_tr:, :]])
tmp = np.ones(crash_data[C_tr:, :].shape[0])
#tmp[:] = 'anomaly'
test_label = np.concatenate([test_label, tmp])

print(train_data.shape, test_data.shape)
print(train_label.shape, test_label.shape)

features = train_data
features

#zdim-norm = 3
umap_2d = UMAP(n_components=2, init='random', random_state=0)
umap_3d = UMAP(n_components=3, init='random', random_state=0)

proj_2d = umap_2d.fit_transform(features)
proj_3d = umap_3d.fit_transform(features)

tmp = np.zeros(train_data.shape[0], dtype="U4")
i = 0
for info in train_label:
  tmp[i] = 'normal' if info == 0 else 'anomaly'
  i+=1

color = dict(zip(range(0,15), plt.cm.tab20(range(0,15))))
fig_2d = px.scatter(proj_2d, x=0, y=1, color = tmp)
fig_3d = px.scatter_3d(proj_3d, x=0, y=1, z=2, color = tmp)
fig_3d.update_traces(marker_size = 2)
fig_2d.show()
fig_3d.show()
```
